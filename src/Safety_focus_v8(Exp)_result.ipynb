# Pulse test results

```python
df_pulse_high = pd.read_excel('../Experiment/Pulse_high.xlsx')

df_pulse_high = df_pulse_high.transpose()
# Set the first row as column names
df_pulse_high.columns = df_pulse_high.iloc[0]

# Drop the first row
df_pulse_high = df_pulse_high.drop(df_pulse_high.index[0])

df_pulse_high = df_pulse_high.astype(float)
df_pulse_high.head()
```

```python
df_pulse_low = pd.read_excel('../Experiment/Pulse_low.xlsx')

df_pulse_low = df_pulse_low.transpose()
# Set the first row as column names
df_pulse_low.columns = df_pulse_low.iloc[0]

# Drop the first row
df_pulse_low = df_pulse_low.drop(df_pulse_low.index[0])

df_pulse_low = df_pulse_low.astype(float)
df_pulse_low.head()
```

```python
df_pulse_med = pd.read_excel('../Experiment/Pulse_medium.xlsx')

df_pulse_med = df_pulse_med.transpose()
# Set the first row as column names
df_pulse_med.columns = df_pulse_med.iloc[0]

# Drop the first row
df_pulse_med = df_pulse_med.drop(df_pulse_med.index[0])

df_pulse_med = df_pulse_med.astype(float)
df_pulse_med.head()
```

```python
plt.rc('font', family='Times New Roman')

# Create a figure and a set of subplots
fig, ax = plt.subplots()

# Plot each column
sns.lineplot(data=df_pulse_high['exogenous shock'], ax=ax, label='high')
sns.lineplot(data=df_pulse_med['exogenous shock'], ax=ax, label='medium',linestyle='--' )
sns.lineplot(data=df_pulse_low['exogenous shock'], ax=ax, label='low', linestyle=':')

# Set the legend
ax.legend()
ax.set_title('exogenous shock')

# Set x and y axis titles
ax.set_xlabel('Time (Month)')
ax.set_ylabel('exogenous shock')

ax.xaxis.tick_top() # This moves the x-axis labels to the top
ax.tick_params(axis='x', direction='in') # This moves the x-axis ticks inside the graph

ax2 = ax.twiny() # This creates a new set of axes with the same y-axis
ax2.xaxis.tick_top() # This moves the new x-axis labels to the top
ax2.set_xticklabels([]) # This removes the new x-axis labels
ax2.tick_params(axis='x', direction='in') # This moves the x-axis ticks inside the graph

# Show the plot
plt.show()
```

```python
plt.rc('font', family='Arial')

# Create a figure and a set of subplots
fig, ax = plt.subplots()

# Plot each column
sns.lineplot(data=df_pulse_high['Attention Resource'], ax=ax, label='Critical Impact Error' )
sns.lineplot(data=df_pulse_med['Attention Resource'], ax=ax, label='Significant Impact Error',linestyle='--' )
sns.lineplot(data=df_pulse_low['Attention Resource'], ax=ax, label='Moderate Impact Error', linestyle=':')

# Set the legend
ax.legend(fontsize = 10)
ax.set_title('Response to Impact of Serious Error to Attention Resource', fontsize=12)

# Set x and y axis titles
ax.set_xlabel('Time (Month)')
ax.set_ylabel('Attention Resource')

ax.xaxis.tick_top() # This moves the x-axis labels to the top
ax.tick_params(axis='x', direction='in') # This moves the x-axis ticks inside the graph

ax2 = ax.twiny() # This creates a new set of axes with the same y-axis
ax2.xaxis.tick_top() # This moves the new x-axis labels to the top
ax2.set_xticklabels([]) # This removes the new x-axis labels
ax2.tick_params(axis='x', direction='in') # This moves the x-axis ticks inside the graph

# Show the plot
plt.show()
# Save the plot
fig.savefig('Images/Pluse_Attention.png')
```

```python
import pysd 
import pandas as pd
import netCDF4
import matplotlib.pyplot as plt
import seaborn as sns 
import numpy as np
```

```python
model = pysd.read_vensim("../Experiment/safety focus_v9(Experiment).mdl")
stocks = model.run()
# model.doc
```

```python
plt.rc('font', family='Arial')
fig, ax = plt.subplots()
stocks["Attention Resource"].plot()
stocks["Organization knowledge"].plot()
plt.title("Firm's Response to Serious Error at Month 20")
plt.ylabel("Attention Resource and Organization knowledge") 
plt.legend()
plt.xlabel("Month")


ax.xaxis.tick_top() # This moves the x-axis labels to the top
ax.tick_params(axis='x', direction='in') # This moves the x-axis ticks inside the graph

ax2 = ax.twiny() # This creates a new set of axes with the same y-axis
ax2.xaxis.tick_top() # This moves the new x-axis labels to the top
ax2.set_xticklabels([]) # This removes the new x-axis labels
ax2.tick_params(axis='x', direction='in') # This moves the x-axis ticks inside the graph

# plt.grid()
plt.show()
```

```python
attention_resource = np.linspace(0,3,20)
organizational_knowledge = np.linspace(0,3, 20)


apv, okv = np.meshgrid(attention_resource, organizational_knowledge)

def derivatives(ar, ok):
    ret  = model.run(params={'attention_resource': ar, 'organization_knowledge': ok}, 
                     return_timestamps=[0, 1], 
                     return_columns=['change_in_attention', 'change_in_knowledge'])
    # print(ret)
    
    return tuple(ret.loc[0].values)

derivatives(0,1)
```

```python
vderivatives = np.vectorize(derivatives)

darv, dokv = vderivatives(apv, okv)
(darv == okv).all()
```

```python
plt.figure(figsize=(5,5))
# plt.quiver(apv, okv, darv, dokv, color='r', alpha=.75)
plt.streamplot(apv, okv, darv, dokv, color='b')

plt.box('off')
# plt.xlim(0,10)
plt.xlabel('Attention Resource')
plt.ylabel('Organization Knowledge')

plt.title('Phase potrait with one shock')
plt.savefig('Images/Streamplot_one_shock.png')
```
